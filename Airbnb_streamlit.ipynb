```python
# Calculate min and max price
min_price = location_df_2['price'].min()
max_price = location_df_2['price'].max()
price_difference = max_price - min_price

# Define price ranges
range_1 = round(min_price + price_difference * 0.30, 2)
range_2 = round(min_price + price_difference * 0.60, 2)

# Create radio button options with formatted price ranges
val_sel = st.radio(
    "Select the Price Range",
    [
        f"${min_price} to ${range_1} (0% to 30% of the Value)",
        f"${range_1} to ${range_2} (30% to 60% of the Value)",
        f"${range_2} to ${max_price} (60% to 100% of the Value)"
    ]
)

# Filter dataframe based on selection
def select_the_df(selected_range):
    if selected_range.startswith(f"${min_price} to ${range_1}"):
        return location_df_2[(location_df_2['price'] >= min_price) & (location_df_2['price'] <= range_1)]
    elif selected_range.startswith(f"${range_1} to ${range_2}"):
        return location_df_2[(location_df_2['price'] > range_1) & (location_df_2['price'] <= range_2)]
    else:
        return location_df_2[(location_df_2['price'] > range_2) & (location_df_2['price'] <= max_price)]

df_val_sel = select_the_df(val_sel)

# Display the filtered dataframe
st.dataframe(df_val_sel)
```
